```python
df = pd.read_csv('train.txt', sep='\t').set_index('id')
```

```python
def concatenate_turn(t1, t2, t3):
    res = [f'{tt1} {tt2} {tt3}' for tt1, tt2, tt3 in zip(list(t1), list(t2), list(t3))]
    return res
```

```python
df.head()
```

```python
df['concatenate'] = concatenate_turn(df.turn1, df.turn2, df.turn3)
```

```python
tokenizer = Tokenizer(num_words=2500, lower=True, split=' ')
tokenizer.fit_on_texts(df.concatenate.values)
X = tokenizer.texts_to_sequences(df.concatenate.values)
X = pad_sequences(X)
```

```python
X
```

```python
embed_dim = 128
lstm_out = 200
batch_size = 32

model = Sequential()
model.add(Embedding(2500, embed_dim, input_length = X.shape[1], dropout = 0.2))
model.add(LSTM(lstm_out, dropout_U = 0.2, dropout_W = 0.2))
model.add(Dense(4,activation='softmax'))
model.compile(loss = 'categorical_crossentropy', optimizer='adam',metrics = ['accuracy'])
print(model.summary())
```

```python
Y = pd.get_dummies(df['label']).values
X_train, X_valid, Y_train, Y_valid = train_test_split(X,Y, test_size = 0.20, random_state = 36)

#Here we train the Network.

model.fit(X_train, Y_train, batch_size =batch_size, nb_epoch = 1,  verbose = 5)
```

```python
score, acc = model.evaluate(X_valid, Y_valid, verbose=2, batch_size=batch_size)
print(f'Score: {score}')
print(f'Val accuracy: {acc}')
```
