```python
turn1 = Input(shape=(100,), dtype='int32', name='main_input1')
turn2 = Input(shape=(100,), dtype='int32', name='main_input2')
turn3 = Input(shape=(100,), dtype='int32', name='main_input3')

emdeddingLayer = Embedding(embeddingMatrix.shape[0], EMBEDDING_DIM,
                          weights=[embeddingMatrix], input_length=MAX_LENGTH, trainable=False)

emb1 = embeddingLaer(turn1)
emb2 = embeddingLaer(turn2)
emb3 = embeddingLaer(turn3)

lstm = Bidirectional(LSTM(LSTM_DIM, dropout=DROPOUT))

lstm1 = lstm(emb1)
lstm2 = lstm(emb2)
lstm3 = lstm(emb3)

inp = Concatenate(axis=1)([lstm1, lstm2, lstm3])

out = Dense(NUM_CLASSES, activation='softmax')(inp)

adam = optimizers.adam(lr=LEARNING_RATE)
model = Model([turn1, turn2, turn3], out)
model.compile(loss='categorical_crossentropy',
              optimizer=adam, metrics=['acc'])

print(model.summary())
```
